```python
from transformers import pipeline

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create machines
that can perform tasks that typically require human intelligence. Examples include
language understanding, visual perception, and decision-making.
"""

summary = summarizer(text, max_length=640, min_length=20, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```
